```python
import json
```

# JSON file upload

```python
fd=open("record_ims.json","r")
txt=fd.read()
fd.close()
record1=json.loads(txt)
```

```python
record1
```

# Add and Modify Items in Inventory

```python
prodID = (input('Enter the Product ID: '))
prodName=(input('Enter the PRoduct Name: '))
Category=(input('Enter the Category Name of the Product: '))
quan = ( input('Enter the Quantity: '))
Price = ( input('Enter the Price: '))
discount=( input('Enter the Discount percentage: '))
```

```python
if(prodID in record1.keys()):
   #to modify item Price, quantity and other attributes if already available in inventory
   record1[str(prodID)]={'Name':prodName,'Category':Category,'Quantity Available':str(int(record1[prodID]['Quantity Available'])+int(quan)),'Price':Price,'Discount':discount}
else:
    #to add new items
   record1[str(prodID)]={'Name':prodName,'Category':Category,'Quantity Available':quan,'Price':Price,'Discount':discount}
```

```python
record1   #Attributes for above entered input has updated
```

# Update JSON file

```python
js=json.dumps(record1)
fd=open("record_ims.json","w")
fd.write(js)
fd.close()
```
